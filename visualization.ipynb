```python
import numpy as np
import matplotlib.pyplot as plt

def plot_admissible_set(admissible_thetas):
    """Plot parameter distributions"""
    param_names = ['m1', 'm2', 'k1', 'k2', 'b1', 'b2']
    
    fig, axes = plt.subplots(2, 3, figsize=(14, 8))
    axes = axes.flatten()
    
    for i, name in enumerate(param_names):
        axes[i].hist(admissible_thetas[:, i], bins=20, alpha=0.7, edgecolor='black', color='gray')
        axes[i].set_xlabel(name, fontsize=12)
        axes[i].set_ylabel('Frequency', fontsize=12)
        axes[i].set_title(f'Distribution of {name}', fontsize=12)
        axes[i].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
```

```python

loaded_array = np.load('admissible_thetas_data.npy')
plot_admissible_set(loaded_array)
```

```python

loaded_array = np.load('admissible_thetas_total.npy')
plot_admissible_set(loaded_array)
```
